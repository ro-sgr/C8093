**1)** Hacer un programa que realice una grafica del valor al que converge la probabilidad de tirar una moneda N veces (hay un resultado para Águila y uno para Sol).

```python
# Caras de la moneda
moneda = ['águila','sol']
```

```python
# Número de veces que se arrojará la moneda
n = int(input("n: "))
```

```python
# Valores en x
x = []
```

```python
# Valores en y
aguila = [] # Lista de probabilidad para águila
sol = [] # Lista de probabilidad para sol
```

```python
a = 0 # Conteo de águila
s = 0 # Conteo de sol

for i in range(1,n+1):
    
    tiro = rnd.choice(moneda)
    
    if tiro == 'águila':
        a += 1
    else:
        s += 1
        
    x.append(i) # Valores en x
        
    pa = a/i # Probabilidad de sacar águila
    aguila.append(pa)
    
    ps = s/i #Probabilidad de sacar sol
    sol.append(ps)

# Si desea correr esta celda nuevamente, corra las anteriores para evitar problemas
```

```python
fig1 = plt.figure()
plt.style.use('seaborn-whitegrid')
ax = fig1.add_axes([0,0,1,1])
ax.scatter(x, aguila, color='red', s=5) # Convergencia de águila
ax.scatter(x, sol, color='blue', s=5) # Convergenia de sol
ax.set_xlabel('\nNúmero de tiros')
ax.set_ylabel('Probabilidad\n')
ax.set_title('\nConvergencia de tirar una moneda\n')
ax.legend(moneda)
plt.rcParams.update({'font.size': 12})
plt.plot();
```

**2)** Hacer un programa que realice  una grafica del valor al que converge la probabilidad de sacar un 3 al tirar un dado de 6 caras N veces 

```python
# Caras del dado
dado = [1,2,3,4,5,6]
```

```python
# Número de veces que se arrojará el dado
n = int(input("n: "))
```

```python
# Valores en x
x = []
```

```python
# Valores en y
tres = [] # Lista de probabilidad de sacar 3
```

```python
t = 0 # Conteo de tres

for i in range(1,n+1):
    
    tiro = rnd.choice(dado)
    
    if tiro == 3:
        t += 1
    else:
        pass
        
    x.append(i) # Valores en x
        
    pt = t/i # Probabilidad de sacar tres
    tres.append(pt)

# Si desea correr esta celda nuevamente, corra las anteriores para evitar problemas
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x, tres, color='green', s=5) # Convergencia de sacar 3
ax.set_xlabel('\nNúmero de tiros')
ax.set_ylabel('Probabilidad\n')
ax.set_title('\nConvergencia de sacar 3\n')
plt.style.use('seaborn-whitegrid')
plt.rcParams.update({'font.size': 6})
plt.plot();
```

**3)** Hacer un programa que realice  una grafica del valor al que converge la probabilidad de obtener como suma 8 al tirar 2 dados de 6 caras N veces.

```python
# Caras de los dados
dado1 = dado
dado2 = dado
```

```python
# Número de veces que se arrojarán los dados
n = int(input("n: "))
```

```python
# Valores en x
x = []
```

```python
# Valores en y
suma_ocho = [] # Lista de probabilidad de sacar 3
```

```python
so = 0 # Conteo de suma ocho

for i in range(1,n+1):
    
    tiro1 = rnd.choice(dado1)
    tiro2 = rnd.choice(dado2)
    
    if tiro1+tiro2 == 8:
        so += 1
    else:
        pass
    
    x.append(i) # Valores en x
        
    pso = so/i # Probabilidad de sumar ocho
    suma_ocho.append(pso)

# Si desea correr esta celda nuevamente, corra las anteriores para evitar problemas
```

```python
fig3 = plt.figure()
plt.style.use('seaborn-whitegrid')
ax = fig3.add_axes([0,0,1,1])
ax.scatter(x, suma_ocho, color='purple', s=5) # Convergencia de sumar 8
ax.set_xlabel('\nNúmero de tiros')
ax.set_ylabel('Probabilidad\n')
ax.set_title('\nConvergencia de sumar 8\n')
plt.rcParams.update({'font.size': 12})
plt.plot();
```
